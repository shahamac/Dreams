# 15.  WHICH AGE GROUP CUSTOMERS SPENT MORE AMOUNT ON SWEET PRODUCTS.

```python
data['Age'].min()
```

```python
data['Age'].max()
```

```python
def grouper(data,idx,col):
    if 26<=data[col].loc[idx]<=40:
        return "26-40"
    elif 41<=data[col].loc[idx]<=50:
        return "41-50"
    elif 51<=data[col].loc[idx]<=60:
        return "51-60"
    elif 61<=data[col].loc[idx]<=70:
        return "61-70"
    elif 71<=data[col].loc[idx]<=82:
        return "71-82"
```

```python
age=data.groupby(lambda x:grouper(data,x,'Age'))
print(age)
```

```python
for key, data in age:
    print("\nAge Group: {}\n".format(key))
    print("Amount spent on sweet products:",data['MntSweetProducts'].sum())
```
